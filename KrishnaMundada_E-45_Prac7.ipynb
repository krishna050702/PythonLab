```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as ny
```

#### Q 1. Read Total profit of all months and show it using a line plot


```python
data =pd.read_csv("company_sales_data.csv")
```

```python
profit =ny.array(data["total_profit"])
print(profit)
```

#### Q 2. Get total profit of all months and show line plot with the following Style properties.
#### Generated line plot must include following Style properties: â€“
#### Line Style dotted and Line-color should be red
#### Show legend at the lower right location.
#### X label name = Month Number
#### Y label name = Sold units number

```python
plt.plot(profit,color= 'r',linestyle ='dashdot',marker='o', linewidth='3')
plt.legend(["Total Profit"],loc= "lower right")
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.show()
```

#### Q-3. Read all product sales data and show it using a multiline plot 
#### Display the number of units sold per month for each product using multiline plots. (i.e.,
#### Separate Plotline for each product).

```python

allunits= ny.array(data['total_units'])
##print(allunits)
facecream = ny.array(data["facecream"])
facewash= ny.array(data["facewash"])
toothpaste = ny.array(data["toothpaste"])
bathsoap= ny.array(data["bathingsoap"])
shampoo= ny.array(data["shampoo"])
moisturizer= ny.array(data["moisturizer"])

plt.plot(facecream, label="facecream")
plt.plot(facewash,label="facewash")
plt.plot(toothpaste,label ="toothpaste ")
plt.plot(bathsoap,label="bathshoap")
plt.plot(shampoo,label=" shampoo")
plt.plot(moisturizer,label="moisturizer")
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.legend()
plt.show()
```

#### Q  8. Calculate total sale data for last year for each product and show it using a Pie chart.


```python
sumprof= ny.sum(profit)
sumcream= (ny.sum(facecream) / sumprof)*100
sumwash=(ny.sum(facewash)/sumprof)*100
sumtoothpaste= (ny.sum(toothpaste)/sumprof)*100
sumsoap = (ny.sum(bathsoap)/sumprof)*100
sumshampoo =(ny.sum(shampoo)/sumprof)*100
summoist=(ny.sum(moisturizer)/sumprof)*100
mylabels = ["facecream","facewash","toothpaste","bathsoap","shampoo","Moisturizer"]
mydata=[]
mydata.append(sumcream)
mydata.append(sumwash)
mydata.append(sumtoothpaste)
mydata.append(sumsoap)
mydata.append(sumshampoo)
mydata.append(summoist)

plt.pie(mydata,labels=mylabels,shadow=True)
plt.show()
```
