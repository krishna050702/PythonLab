Task 1: Develop a Memory Game using pygames and display users score.\
Display 4 to 5 images at random for some small time duration\
For example,\
display the images of different numbers:\
1\
3\
4\
2\
1\
4\
5\
2\
3\
2\
After the display of every image screen should become blank.\
you can define the total time for playing the game.\
After watching the images, ask the user to tell how many images of each number were shown\
and display the score accordingly.\
Expected answer from user:\
1 is displayed 2 times\
3 is displayed 2 times\
2 is displayed 3 times\
4 is displayed 2 times

```python
import pygame,sys,random,os
from pygame.locals import *
pygame.init()

# set up the colors
BLACK = (0,0,0)
WHITE = (255, 255, 255)

screen = pygame.display.set_mode([500, 500])
myFont = pygame.font.SysFont("Times New Roman", 100)
myFont2 = pygame.font.SysFont("Times New Roman", 20)
path = r"C:\Users\Intel\Python\NumImgs"

pygame.display.set_caption('MEMORY GAME')

dictnum = {}
for i in range(0,8):
    randomNum = random.choice([
        x for x in os.listdir(path)
        if os.path.isfile(os.path.join(path, x))
    ])
    numImage = pygame.image.load(path+"\\" +randomNum)
    num = randomNum.split(".")[0]
    num = int(num)
    
    if num in dictnum:
        dictnum[num]+=1
    else:
        dictnum[num]=1

    screen.blit(numImage, (50, 50))
    pygame.display.update()
    pygame.time.delay(1500)
    screen.fill(BLACK)
    pygame.display.update()

    
prompt = myFont2.render(str("How many times did each number appear?"), 1, WHITE)
screen.blit(prompt, (10, 10))
pygame.display.update()

 # run the game loop
while True:
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
    pygame.display.update()
```

```python
score=0
for i in range(len(dictnum)):
    answer = input('Enter your answer: Number<space>Occurences:  ')
    ans=answer.split()
    key=int(ans[0])
    val=int(ans[1])
    if (key in dictnum and dictnum[key]==val):
        score+=val

print("FINAL SCORE:  ", score, "/",8)
print("_______________________")
print("CORRECT ANSWERS-")
for e,v in dictnum.items():
    print(e,": ",v)
```

Task 2: Consider Names of two places from the user and find out the distance between them
using Geopy library and Mark the location using folium

```python
from geopy.geocoders import Nominatim as nm
from geopy.distance import geodesic as GD 
import folium
place1 = input("Enter first location: ")
place2 = input("Enter second location: ")
geoloc = nm(user_agent="my_request")
loc1 = (geoloc.geocode(place1).latitude,geoloc.geocode(place1).longitude)
loc2 = (geoloc.geocode(place2).latitude,geoloc.geocode(place2).longitude) 
print(loc1)
print(loc2)
world = folium.Map()
folium.Marker(loc1,).add_to(world)
folium.Marker(loc2,).add_to(world)
print(f"The distance between {place1} and {place2} is:  {GD(loc1, loc2).km} KM")
world
```

Task 3. Using “list-indian-states-and-capitals.csv” file get the name of Indian-state-capitals
using Geopy find the latitude and longitude of the Indian-state-capitals and label the capital on
the map using folium.

```python
import pandas as pd
```

```python
file = "list-indian-states-and-capitals.csv"
df=pd.read_csv(file).dropna()
new_header = df.iloc[0]
df = df[1:] 
df.columns = new_header 
df
```

```python
from geopy.geocoders import Nominatim as nm
from geopy.distance import geodesic as GD 
import folium
world = folium.Map()
geoloc = nm(user_agent="my_request")
count=0

for cities in df['Capital']:
    for city in cities.split("\n"):
        city = city.split('(')[0]
        count+=1
        loc = (geoloc.geocode(city).latitude,geoloc.geocode(city).longitude)
        print(f"{count}. {city}: {loc}")
        folium.Marker(loc).add_to(world)
world    
```
